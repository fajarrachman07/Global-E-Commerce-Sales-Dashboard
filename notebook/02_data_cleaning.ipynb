# I. Data Loading

Memuat dataset transaksi e-commerce sebagai data mentah sebelum dilakukan data cleaning dan proses transformasi data

```python
import pandas as pd

df = pd.read_csv("../data/global_ecommerce_sales.csv")

df.head()
```

Dataset berhasil dimuat sebagai data mentah yang akan diproses lebih lanjut melalui tahapan transformasi untuk memastikan kesiapan data sebelum dilakukan exploratory data analysis dan visualisasi dashboard.

# II. Convert order_date to Datetime

Mengubah tipe data kolom Order_Date dari object menjadi datetime agar dapat digunakan dalam analisis
berbasis waktu seperti tren penjualan bulanan, tahunan, maupun identifikasi pola musiman transaksi.

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
```

```python
df.info()
```

Kolom Order_Date berhasil dikonversi dari tipe data object menjadi datetime, sehingga memungkinkan analisis tren
penjualan berbasis waktu seperti perbandingan performa penjualan antar bulan ataupun tahun.

Transformasi ini menjadi penting sebelum melakukan exploratory data analysis pada pola transaksi temporal.

# III. Create Time Based Features

Membuat variabel turunan berbasis waktu untuk mendukung analisis tren penjualan secara temporal.

```python
df["Year"] = df["Order_Date"].dt.year
df["Month"] = df["Order_Date"].dt.month
df["Month_Name"] = df["Order_Date"].dt.month_name()
```

```python
df.head()
```

Kolom Year, Month, dan Month_Name berhasil dibuat untuk mendukung analisis tren penjualan berdasarkan waktu serta mempermudah proses agregasi data pada dashboard interaktif.

# IV. Create Profit_Margin Feature

Menghitung rasio profit terhadap total penjualan sebagai indikator efisiensi profitabilitas transaksi.

```python
df["Profit_Margin"] = df["Profit"] / df["Total_Sales"]
```

```python
df.head()
```

Kolom Profit_Margin menunjukkan rasio keuntungan terhadap total penjualan pada setiap transaksi. Nilai ini digunakan untuk mengukur efisiensi profitabilitas penjualan, sehingga memungkinkan identifikasi kategori produk, region, atau segmen pelanggan yang menghasilkan margin keuntungan lebih tinggi dibandingkan yang lain.

# V. Identify Negative Profit Transactions

Mengidentifikasi transaksi dengan nilai profit negatif untuk memahami potensi faktor yang menyebabkan kerugian pada penjualan.

```python
# Melihat Transaksi Yang Rugi

df[df["Profit"] < 0]
```

```python
# Menhitung Jumlah Transaksi Rugi

(df["Profit"] < 0).sum()
```

```python
# Menghitung Proporsinya

(df["Profit"] <0).mean()
```

```python
df[df["Profit"] < 0][["Product_Category", "Discount_Percent", "Shipping_Cost"]]
```

Ditemukan sebanyak 272 transaksi atau sekitar 13.6% dari total transaksi menghasilkan profit negatif. Sebagian besar transaksi yang merugi berasal dari kategori Office Supplies, yang menunjukkan kemungkinan rendahnya margin produk pada kategori tersebut. Selain itu, nilai Shipping_Cost yang relatif tinggi dibandingkan Total_Sales serta penerapan diskon pada beberapa transaksi diduga menjadi faktor utama yang berkontribusi terhadap kerugian penjualan. Temuan ini menunjukkan adanya potensi ketidakefisienan pada strategi pricing atau distribusi untuk kategori produk tertentu yang perlu dianalisis lebih lanjut pada tahap exploratory data analysis.

Memastikan dataset telah melalui proses transformasi yang diperlukan dan siap digunakan pada tahap
